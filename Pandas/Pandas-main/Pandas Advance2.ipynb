#### Merge
Merge in pandas is done in two ways hrizontally or Vertically

```python
data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df1 = data[["Name","Survived","Pclass"]][0:5]
```

```python
df2 = data[["Name","Survived","Pclass"]][5:10]
```

```python
## by default pandas merges vertically that is axis = 0
pd.concat([df1,df2])
```

```python
df3 = pd.concat([df1,df2],axis = 1)
df3
```

```python
## fill na indf 3
df3.fillna("Prakash")
```

```python
data1 = pd.DataFrame({"key1" : [1,4,2,34],
                      "key2" : [5,4,6,78],
                      "key3" : [9,1,7,32]})

data1 
```

```python
data2 = pd.DataFrame({"key1" : [1,20,2,34],
                      "key4" : [5,4,6,780],
                      "key5" : [9,1,70,32]})
data2
```

```python
## inner join by default
pd.merge(data1,data2)        
```

```python
## left join 
pd.merge(data1,data2,how='left')
```

```python
## right join
## we can use on

# pd.merge(data1,data2,how = "right",left_on = 'key2',right_on = 'key4')

pd.merge(data1,data2,how = "right",left_on = 'key2',right_on = 'key4')
```

```python
## outer join
pd.merge(data1,data2, how = 'outer')
```

```python
## cross join (row 4*4) column(3*3)
## every row has a match
pd.merge(data1,data2, how = 'cross')
```

##### Join operations
join works on indexes

```python
data2 = pd.DataFrame({"key1" : [1,20,2,34],
                      "key4" : [5,4,6,780],
                      "key5" : [9,1,70,32]},index= ['a','b','c','d'])
```

```python
data1 = pd.DataFrame({"key11" : [1,4,2,34],
                      "key12" : [5,4,6,78],
                      "key13" : [9,1,7,32]},index = ['a','b','g','h'])
```

```python
## by default it is taking left join
data1.join(data2)
```

```python
data1.join(data2,how="right")
```

##### Create another column where below 10 euro == cheap 10-20 medium and above20 == high

```python

def category_fare(x):
        
        if x < 10 :
            return "cheap"
        elif x> 10 and x<20 :
            return "medium"
        else :
            return "high"
```

```python
data["cat_fare"] = data["Fare"].apply(category_fare)
```
